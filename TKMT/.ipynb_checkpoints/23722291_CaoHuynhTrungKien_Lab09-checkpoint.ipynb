### Bài 1

Nghiên cứu mức thu nhập ($X$) và chi tiêu ($Y$) trong một tháng của một công ty. Khảo sát ngẫu nhiên 7 nhân viên, ta thu được bảng số liệu sau:

| X            | 130     | 150      | 133      | 170      | 170      | 210      | 230      |
| :----------: | :-----: | :------: | :------: | :------: | :------: | :------: | :------: |
| **Y**        | **84**  | **120**  | **108**  | **130**  | **130**  | **150**  | **160**  |


a. Dựa vào bảng dữ liệu trên cho biết có xây dựng được mô hình hồi quy hay không? Nếu có hãy xây dựng mô hình hồi quy để ước tính chi tiêu dựa theo thu nhập

b. Dự đoán mức chi tiêu của nhân viên công ty trên nếu biết thu nhập là 200$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dữ liệu thu nhập (X) và chi tiêu (Y)
X = np.array([130, 150, 133, 170, 170, 210, 230]).reshape(-1, 1)
Y = np.array([84, 120, 108, 130, 130, 150, 160])

# a. Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X, Y)

# Lấy hệ số hồi quy
a = model.intercept_
b = model.coef_[0]

print(f"Hàm hồi quy: Y = {a:.2f} + {b:.2f} * X")

# Vẽ biểu đồ
plt.scatter(X, Y, color='blue', label='Dữ liệu thực tế')
plt.plot(X, model.predict(X), color='red', label='Đường hồi quy')
plt.xlabel("Thu nhập ($)")
plt.ylabel("Chi tiêu ($)")
plt.title("Hồi quy tuyến tính giữa Thu nhập và Chi tiêu")
plt.legend()
plt.grid(True)
plt.show()

# b. Dự đoán chi tiêu nếu thu nhập là 200$
thu_nhap_moi = np.array([[200]])
chi_tieu_du_doan = model.predict(thu_nhap_moi)
print(f"Dự đoán chi tiêu khi thu nhập là 200$: {chi_tieu_du_doan[0]:.2f} $")
```

### Bài 2

Tiến hành nghiên cứu mối quan hệ giữa:

- X: giá bán kẻ của thịt gà (nghìn/kg) 
- Y: lượng thịt gà tiêu thụ (kg/tháng)

Dữ liệu thu được gồm 10 tháng như sau:

| X            | 35       | 40       | 45       | 40       | 38       | 40       | 45       | 30       | 42       | 38       |
| :----------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: |
| **Y**        | **2.29** | **2.04** | **0.59** | **2.06** | **2.50** | **2.32** | **1.57** | **3.96** | **1.43** | **2.34** |


a. Dựa vào bảng dữ liệu trên cho biết có xây dựng được mô hình hồi quy hay không? Nếu có hãy xây dựng mô hình hồi quy để ước tính lượng tiêu thụ thịt gà theo giá bán.

b. Dự đoán lượng tiêu thụ khi giá bán lẻ là 41(kg/tháng)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dữ liệu: X = giá bán thịt gà (nghìn/kg), Y = lượng tiêu thụ (kg/tháng)
X = np.array([35, 40, 45, 40, 38, 40, 45, 30, 42, 38]).reshape(-1, 1)
Y = np.array([2.29, 2.04, 0.59, 2.06, 2.50, 2.32, 1.57, 3.96, 1.43, 2.34])

# a. Xây dựng mô hình hồi quy
model = LinearRegression()
model.fit(X, Y)

# Lấy hệ số hồi quy
a = model.intercept_
b = model.coef_[0]

print(f"Hàm hồi quy: Y = {a:.4f} + ({b:.4f}) * X")

# Vẽ biểu đồ
plt.scatter(X, Y, color='green', label='Dữ liệu thực tế')
plt.plot(X, model.predict(X), color='red', label='Đường hồi quy')
plt.xlabel("Giá bán thịt gà (nghìn đồng/kg)")
plt.ylabel("Lượng tiêu thụ (kg/tháng)")
plt.title("Hồi quy giữa giá bán và lượng tiêu thụ thịt gà")
plt.legend()
plt.grid(True)
plt.show()

# b. Dự đoán lượng tiêu thụ khi giá bán là 41 (nghìn/kg)
gia_ban_moi = np.array([[41]])
luong_tieu_thu_du_doan = model.predict(gia_ban_moi)
print(f"Dự đoán lượng tiêu thụ khi giá bán 41 nghìn/kg: {luong_tieu_thu_du_doan[0]:.2f} kg/tháng")
```

## Bài 3

Dụa trên file dữ liệu `Advertising.csv` bạn hãy dự đoán doanh số bán hàng (`sales`) dựa trên số tiền chi cho các nền tảng tiếp thị khác nhau như `TV`, `Radio` và `Newspaper`

a. Đọc file dữ liệu cho biết các thông tin cơ bản về dữ liệu: kích thước, dữ liệu có giá trị trống hay không, hiện các giá trị thống kê cơ bản của các thuộc tính.

b. Để dự đoán doanh số bằng phương pháp hồi quy tuyến tính đơn giản (Simple Linear Regression) trước tiên cần lựa chọn một trong các thuộc tính còn lại. Theo bạn thuộc tính nào phù hợp để xây dựng mô hình hồi quy. Giải thích và minh họa bằng 2 cách

c. Xây dụng phương trình hồi quy, trực quan mô hình (đường thẳng) tìm được. Theo bạn để đánh giá mô hình trên cần dùng đại lượng nào?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc file CSV (nếu bạn chưa có file thì cần upload trước)
df = pd.read_csv("Advertising.csv")

# Hiển thị kích thước
print("Kích thước dữ liệu:", df.shape)

# Kiểm tra giá trị null
print("\nCó giá trị trống không?")
print(df.isnull().sum())

# Thống kê mô tả
print("\nThông tin thống kê mô tả:")
print(df.describe())

#CÂU B

# Tính hệ số tương quan Pearson
corr = df.corr()
print("\nHệ số tương quan:")
print(corr['sales'].sort_values(ascending=False))

# Trực quan hoá
sns.pairplot(df, x_vars=['TV', 'radio', 'newspaper'], y_vars='sales', height=5, aspect=0.8, kind='reg')
plt.suptitle("Mối quan hệ giữa các kênh quảng cáo và doanh số", fontsize=14)
plt.show()

#CÂU C
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Xây mô hình hồi quy đơn với TV
X = df[['TV']]
y = df['sales']

model = LinearRegression()
model.fit(X, y)

# Hệ số hồi quy
a = model.intercept_
b = model.coef_[0]
print(f"\nPhương trình hồi quy: Sales = {a:.4f} + {b:.4f} * TV")

# Trực quan hoá
plt.scatter(X, y, color='blue', label='Dữ liệu thực tế')
plt.plot(X, model.predict(X), color='red', label='Hồi quy tuyến tính')
plt.xlabel("Chi phí quảng cáo trên TV")
plt.ylabel("Doanh số bán hàng")
plt.title("Hồi quy tuyến tính: TV vs Sales")
plt.legend()
plt.grid(True)
plt.show()

# Đánh giá mô hình
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print(f"Đánh giá mô hình:")
print(f" - R^2 (hệ số xác định): {r2:.4f}")
print(f" - MSE (Mean Squared Error): {mse:.4f}")


```
